```python
# rereading the file with 'Not Available' and 'unknown' as null values.

df = pd.read_csv("RetailOrders/orders.csv", na_values = ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
df[df['Ship Mode'].isnull()]
```

```python
# renaming the columns to all lower case

df.columns = df.columns.str.lower()
df.columns
```

```python
# replacing " " in column names with " _ "

df.columns = df.columns.str.replace(" ", "_") 
df.columns
```

```python
df['discount'] = df['list_price'] * df['discount_percent']* .01
df.head()
```

```python
df['sale_price'] = df['list_price'] - df['discount']
df.head()
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace = True)
```
